#### Задание 5. Используя pandas, проведи когортный анализ пользователей.
В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.

```python
#объединим таблицу с id пользователей и таблицу с заказами
join_id_orders= uniq_id_cust.merge(orders, how = 'outer', on ='customer_id')
```

```python
join_id_orders.head ()
```

```python
# приведем к формуту даты
join_id_orders ['order_approved_at'] = pd.to_datetime(join_id_orders ['order_approved_at'])
```

```python
# перевели в год, в котором сделан заказ и оплачен
join_id_orders ['year_approved'] = join_id_orders.order_approved_at.dt.year
```

#### В данных у нас полный год только один - 2017. Поэтому в исследовании будем рассматривать только его.

```python
# рассмотрим 2017 год
join_id_orders = join_id_orders.query ('year_approved == "2017.0"')
```

```python
join_id_orders.head ()
```

```python
# Уберем ненужные для анализа колонки
join_id_orders.drop(columns=['customer_id','customer_zip_code_prefix', 'customer_city', 'customer_state',
                            'order_purchase_timestamp', 'order_delivered_carrier_date', 'order_delivered_customer_date',
                            'order_estimated_delivery_date', 'year_approved'], inplace=True)
```

```python
join_id_orders.head ()
```

```python
# Найдем месяц самого первого заказа по клиентам
# Этот же месяц буде твляться искомой когортой (в дальнейшем туда войдут те, кто задержится на 3 месяца)
join_id_orders['Kog_min_month'] = join_id_orders.groupby('customer_unique_id')['order_approved_at'].transform('min').dt.month
```

```python
# Найдем месяц самого последнего заказа по клиентам
join_id_orders['max_month'] = join_id_orders.groupby('customer_unique_id')['order_approved_at'].transform('max').dt.month
```

```python
join_id_orders.head ()
```

#### На основе первого и последнего месяца мы сможем отобрать клиентов, которые оставались в течении трех первых месяцев с момента их первой покупки

```python
# Отобрали только тех, кто еще делал заказы в течении первых трех месяцев смомента покупки.
sub = join_id_orders\
    .query ('Kog_min_month + 2 <= max_month')
```

```python

sub
```

```python
'''
Так как в ноябре и декабре не будет тех, кто останется на третий месяц (не хватит кол-ва месяцев в году), 
мы сделаем условие на ограничение месяцев до октября включительно
'''

sub_count = sub\
    .query ('Kog_min_month < 11')\
    .groupby ('Kog_min_month', as_index=False)\
    .agg ({'customer_unique_id': 'count'})\
    .rename(columns = {'Kog_min_month' : 'kohort', 'customer_unique_id' : 'count_3_month'}) 
```

```python
sub_count
```

```python
# И посчитаем общее кол-во клиентов, которые делали заказ в каждом месяце
all_count = join_id_orders\
    .groupby ('Kog_min_month', as_index=False)\
    .agg ({'customer_unique_id': 'count'})\
    .rename(columns = {'Kog_min_month' : 'kohort', 'customer_unique_id' : 'count_all'}) 
```

```python
all_count
```

```python
#объединим таблицы с id пользователей и таблицу с заказами
Retention= all_count.merge(sub_count, how = 'outer', on ='kohort')
```

```python
Retention = Retention.query ('kohort <11')
Retention
```

```python
Retention ['Ret'] = (Retention['count_3_month']/ Retention ['count_all']) * 100
```
